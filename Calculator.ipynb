```python
# Run this in Jupyter Notebook or as .py file
%gui tk

import tkinter as tk
from math import *

# Main window
root = tk.Tk()
root.title("Scientific Calculator")
root.geometry("400x600")
root.configure(bg="#2b2b2b")
root.resizable(False, False)

expression = ""

# Function to update expression
def press(key):
    global expression
    expression += str(key)
    equation.set(expression)

# Function to evaluate the final expression
def equalpress():
    global expression
    try:
        total = str(eval(expression))
        equation.set(total)
        expression = total
    except:
        equation.set("Error")
        expression = ""

# Function to clear the screen
def clear():
    global expression
    expression = ""
    equation.set("")

# Create the equation text
equation = tk.StringVar()

# Entry widget
expression_field = tk.Entry(root, textvariable=equation, font=('Arial', 20, 'bold'), bg="#1e1e1e", fg="white", bd=10, relief="ridge", justify="right")
expression_field.grid(row=0, column=0, columnspan=5, ipadx=8, ipady=25, pady=10)

# Button Design
button_style = {'font': ('Arial', 14), 'bd': 5, 'fg': 'white', 'bg': '#4a4a4a', 'width': 5, 'height': 2}

# Buttons Layout
buttons = [
    ('7', 1, 0), ('8', 1, 1), ('9', 1, 2), ('/', 1, 3), ('sqrt(', 1, 4),
    ('4', 2, 0), ('5', 2, 1), ('6', 2, 2), ('*', 2, 3), ('**', 2, 4),
    ('1', 3, 0), ('2', 3, 1), ('3', 3, 2), ('-', 3, 3), ('log(', 3, 4),
    ('0', 4, 0), ('.', 4, 1), ('%', 4, 2), ('+', 4, 3), ('exp(', 4, 4),
    ('sin(', 5, 0), ('cos(', 5, 1), ('tan(', 5, 2), ('(', 5, 3), (')', 5, 4),
    ('Clear', 6, 0), ('=', 6, 1, 4)
]

# Creating all buttons
for (text, row, col, colspan) in [(btn[0], btn[1], btn[2], 1) if len(btn) == 3 else btn for btn in buttons]:
    action = lambda x=text: press(x) if x not in ['Clear', '='] else (clear() if x == 'Clear' else equalpress())
    b = tk.Button(root, text=text, command=action, **button_style)
    b.grid(row=row, column=col, columnspan=colspan, padx=5, pady=5, sticky="nsew")

root.mainloop()
```
